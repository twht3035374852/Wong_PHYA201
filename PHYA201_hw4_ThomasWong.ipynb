```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
A = 1 # constant (with units) normalize to 1 for simplicity
theta = np.linspace(0, 2*np.pi, 1000)  # uniform distribution of theta
phi= 0  # looking at phi=0

# Define beta as a function of gamma
def beta(gamma_):
    return np.sqrt(1-1/gamma_**2)

# Define dP/dOmega parallel as a function of gamma and theta
def dPparallel(gamma_,theta_):
    return A/(1-beta(gamma_)*np.cos(theta_))**4 * (np.sin(theta_)/(1-beta(gamma_)*np.cos(theta_)))**2

# Define dP/dOmega perpendicular as a function of gamma and theta
def dPperp(gamma_,theta_):
    return A/(1-beta(gamma_)*np.cos(theta_))**4*(1 - (np.sin(theta_)*np.cos(phi)/gamma_/(1-beta(gamma_)*np.cos(theta_)))**2)
```

```python
plt.figure(figsize=(8, 6))
ax = plt.subplot(111, polar=True)  # using polar coordinates since we are looking at a slice of phi=0

ax.plot(theta,dPparallel(1,theta), 'tab:green', label=r'$\gamma=1$')
ax.set_title('Radial Plot with Radius Varying as a Function of Polar Angle')
ax.set_title(r'$\rm{Parallel \ Angular \ Distribution \ of \ Radiation}$', size=20)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)
ax.legend(fontsize=15, loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

### Discussion
For the parallel angular power distribution, the magnitude does not differ by that much so they can all be plotted on the same polar plot; but for the perpendicular angular power distribution, the magnitude ranges about nine orders of magnitude (plotting $\gamma=1.01$ to illustrate the tiny blob behind).  If we plot the radial direction (which is $dP/d\Omega$) in $\log_{10}$ scale, the small blob behind would be too dominant as seen on the graph as log scale can go to negative to represent how small the blob is.
